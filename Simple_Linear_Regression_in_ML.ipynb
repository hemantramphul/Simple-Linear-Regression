# **Simple Linear Regression**

### Importing the libraries

```python
import numpy as np  # NumPy for numerical operations
import matplotlib.pyplot as plt  # Matplotlib for plotting
import pandas as pd  # Pandas for data manipulation
import statsmodels.api as sm  # Statsmodels for statistical modeling
```

### Importing the dataset

```python
# Import a CSV file using the read_csv() function from the pandas library.
# Read data from a URL with the pandas.read_csv()

# Load the dataset from the specified CSV file
data_set= pd.read_csv('../content/sample_data/salary_data.csv')

# Create a scatter plot using the 'YearsExperience' as x-axis and 'Salary' as y-axis
data_set.plot(x="YearsExperience", y="Salary", kind="scatter")
```

```python
# Create the linear regression model
ey = data_set['Salary']  # Dependent variable
ex = data_set['YearsExperience']  # Independent variable
ex = sm.add_constant(ex)  # Adding a constant term for the intercept
linear_model = sm.OLS(ey, ex)  # Ordinary Least Squares (OLS) method for linear regression

# Fit the linear regression model
lin_result = linear_model.fit()

# Print the summary of regression results
print(lin_result.summary())
```

```python
# Extract independent variable (X) and dependent variable (y) from the dataset
x = data_set.iloc[:, :-1].values  # Independent variable (all columns except the last one)
y = data_set.iloc[:, 1].values  # Dependent variable (second column)

# Uncomment the line below to print the values of the dependent variable (y)
# print(y)
```

### Splitting the dataset into the Training set and Test set

```python
# Import the train_test_split function from sklearn
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)
```

### Training the Simple Linear Regression model on the Training set

```python
# Import the LinearRegression class from sklearn
from sklearn.linear_model import LinearRegression

# Create a linear regression model
regressor = LinearRegression()

# Fit the model using the training data
regressor.fit(x_train, y_train)
```

### Predicting the Train and Test set results


```python
# Use the trained model to make predictions on the test set
y_pred = regressor.predict(x_test)

# Use the trained model to make predictions on the training set (optional)
x_pred = regressor.predict(x_train)
```

### Visualising the Training set results

```python
# Plot the training data points and the linear regression line
plt.scatter(x_train, y_train, color="green", label="Actual Data")
plt.plot(x_train, x_pred, color="violet", label="Linear Regression Line")
plt.title("Salary vs Experience (Training Dataset)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary (in MUR)")
plt.legend()  # Show legend for clarity
plt.show()
```

### Visualising the Test set results

```python
# Plot the test data points and the linear regression line
plt.scatter(x_test, y_test, color="brown", label="Actual Test Data")
plt.plot(x_train, x_pred, color="skyblue", label="Linear Regression Line")
plt.title("Salary vs Experience (Test Dataset)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary (in MUR)")
plt.legend()  # Show legend for clarity
plt.show()
```
